```python
from src.args.yaml_config import YamlConfig
from src.datasets.brain2text import Brain2TextDataset
from src.args.base_args import B2TDatasetArgsModel
```

```python
import matplotlib.pyplot as plt
```

```python
# Sentence amount per split for test, train and validation
# competition_mode = False
train_ds = Brain2TextDataset(B2TDatasetArgsModel(), YamlConfig().config, split="train")
val_ds = Brain2TextDataset(B2TDatasetArgsModel(), YamlConfig().config, split="val")
test_ds = Brain2TextDataset(B2TDatasetArgsModel(), YamlConfig().config, split="test")

print(len(train_ds))
print(len(val_ds))
print(len(test_ds))
labels = ["Train", "Validation", "Test"]
plt.bar(
    labels,
    [len(train_ds), len(val_ds), len(test_ds)],
    color=["blue", "orange", "green"],
)

# Adding labels and title
plt.ylabel("Amount of sentences")
plt.title("Number of sentences for Train, Test, and Validation Set")

# Display the plot
plt.show()
```

```python
# Sentence amount per split for test, train and validation
# competition_mode = True
train_ds = Brain2TextDataset(
    B2TDatasetArgsModel(competition_mode=True), YamlConfig().config, split="train"
)
val_ds = Brain2TextDataset(
    B2TDatasetArgsModel(competition_mode=True), YamlConfig().config, split="val"
)
test_ds = Brain2TextDataset(
    B2TDatasetArgsModel(competition_mode=True), YamlConfig().config, split="test"
)

print(len(train_ds))
print(len(val_ds))
print(len(test_ds))
labels = ["Train", "Validation", "Test (Competition Holdout)"]
plt.bar(
    labels,
    [len(train_ds), len(val_ds), len(test_ds)],
    color=["blue", "orange", "green"],
)

# Adding labels and title
plt.ylabel("Amount of sentences")
plt.title("Number of sentences for Train, Test, and Validation Set in competition mode")

# Display the plot
plt.show()
```
